```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import common

common.pltsetup()
plt.rcParams.update({"figure.figsize": [4.5, 3.4]})


DATASET = Path('data/topology/clusteringcoeff')


def show_distribution(name):
    title = "Clustering coefficient distribution ({}{})".format(
        common.LAYERS[name],
        ' â€” 1% uniform sample'
    )
    
    x, y = common.load_text_distribution(DATASET / 'distribution-{}.txt'.format(name.replace('+', '')))
    dist = common.Distribution(x, y, "Number of closed triangles in neighborhood", "Number of nodes", title)
    common.plot_all(dist, '../img/topology/clusteringcoeff', name)
```

```python
show_distribution('full')
show_distribution('dir+cnt')
show_distribution('rev')
show_distribution('rel+rev')
show_distribution('ori+snp')
```
