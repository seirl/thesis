```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import common

common.pltsetup()
plt.rcParams.update({"figure.figsize": [4.5, 3.4]})



DATASET = Path('data/topology/connectedcomponents')


def show_distribution(name):
    title = "Connected components ({})".format(common.LAYERS[name])
    
    x, y = common.load_text_distribution(DATASET / name / 'distribution.txt')
    dist = common.Distribution(x, y, "Size of each component", "Number of components", title)
    print("{}: number of CC = {},  size largest = {},  percent of nodes in largest = {}, percent <100 node = {}".format(
        name,
        sum(y),
        x[-1],
        x[-1] / sum(degree * count for degree, count in zip(x, y)) * 100,
        sum(y[0:100]) / sum(degree * count for degree, count in zip(x, y)) * 100,
    ))
    common.plot_all(dist, '../img/topology/connectedcomponents', name)
```

```python
show_distribution('full')
show_distribution('dir+cnt')
show_distribution('rev')
show_distribution('rel+rev')
show_distribution('ori+snp')
```

## By number of origins

```python
def show_distribution_by_origin(name):
    title = "Connected components by number of origins ({})".format(name)
    
    x, y = common.load_text_distribution(DATASET / 'by_origins' / name / 'distribution.txt')
    dist = common.Distribution(x, y, "Number of origins in each component", "Number of components", title)
    print("{}: number of CC = {},  size largest = {},  percent of origins in largest = {}, percent <100 origins = {}".format(
        name,
        sum(y),
        x[-1],
        x[-1] / sum(degree * count for degree, count in zip(x, y)) * 100,
        sum(y[0:100]) / sum(degree * count for degree, count in zip(x, y)) * 100,
    ))
    common.plot_all(dist, '../img/topology/connectedcomponents/by_origins', name)
```

```python
show_distribution_by_origin('full')
show_distribution_by_origin('ori+snp+rel+rev')
show_distribution_by_origin('ori+snp')
```
