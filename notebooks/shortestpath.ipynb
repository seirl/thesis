```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import common

common.pltsetup()
plt.rcParams.update({"figure.figsize": [4.5, 3.4]})

DATASET = Path('data/topology/shortestpath')


def show_distribution(name):
    title = "Path length distribution ({}{})".format(
        common.LAYERS[name],
        ' â€” 15% uniform sample' if name == 'dir+cnt' else ''
    )
    
    x, y = common.load_text_distribution(DATASET / name / 'distribution.txt')
    dist = common.Distribution(x, y, "Size of each shortest path", "Number of paths", title)
    common.plot_all(dist, '../img/topology/shortestpath', name)
```

```python
show_distribution('dir+cnt')
show_distribution('rev')
```
